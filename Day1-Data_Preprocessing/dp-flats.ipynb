```python
# data overview
df.info()
```

```python
# duplicate check
df.duplicated().sum()
```

```python
# null check
df.isnull().sum()
```

```python
# dropping irrelevant columns
df.drop(columns=['link'], inplace=True)
df.head()
```

```python
# in area col it's price per square feet
df.rename(columns={'area':'price_per_sqft'}, inplace=True)
df.head()
```

```python
# check society col
df['society'].value_counts()
```

society names are duplicated with lowercasing and uppercasing, then some got reviews attached to them need to check it.

```python
# there are total 638 unique societies, need to reduce them to handle them properly
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower() # remvoving star marks in society names
```

```python
df['society'].value_counts().shape
```

```python
# on to the price col
df.head()
```

```python
df['price'].value_counts()
```

```python
# there are some values with "price on request", removing them
df[df['price'] != 'Price on Request'].shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
# some prices are in lakhs and some are in crores, let's convert all to crores + scaling down
def convert_to_crores(price):
    if type(price) == float:
        return price
    else:
        if price[1] == 'Lac':
            return round(float(price[0]) / 100, 2)
        else: 
            return round(float(price[0]), 2)
```

```python
df['price'] = df['price'].str.split(" ").apply(convert_to_crores)
```

```python
df.head()
```

```python
# onto price per sqft col, has rs. sign-> no need so removing it
df['price_per_sqft'] = df['price_per_sqft'].str.split("/").str.get(0).str.replace("₹", "").str.replace(",", "").str.strip().astype('float')
```

```python
df.head()
```

```python
# areaWithType col has many issues, needs to be handled properly
df['areaWithType'].value_counts()
```

```python
# col - bedroom, bathroom, balcony have similar issues, so handled them together
df['bedRoom'].value_counts()
```

```python
# null check
df['bedRoom'].isnull().sum()
```

```python
# there are few rows with null values, romved them
df = df[df['bedRoom'].notnull()]
```

```python
df['bedRoom'] = df['bedRoom'].str.split(" ").str.get(0).astype('int')
```

```python
df.head()
```

```python
# same for bathroom and balcony
df['bathroom'] = df['bathroom'].str.split(" ").str.get(0).astype('int')
```

```python
df.head()
```

```python
df['balcony'] = df['balcony'].str.split(" ").str.get(0).str.replace("No", '0')
```

```python
df.head()
```

```python
df['additionalRoom'].value_counts()
```

```python
# unable to think of a way to handle this, so keeping as it is with basic cleaning
df['additionalRoom'].fillna('not available', inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

```python
# floor num col has 2 info - floor number and total floors
# using only 1st info, because another dataset doesn't have 2nd info in it.
df['floorNum'].value_counts()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace("Ground", '0').str.replace("Basement", '-1').str.replace('Lower', '0').str.extract(r'(\d+)')
```

```python
df.head()
```

```python
# facing col has many missing values, replacing them "NA"
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA', inplace=True)
df['facing'].value_counts()
```

```python
# created new col = 'area' using price and price_per_sqft, added at loc = 4
df.insert(loc=4, column='area', value=round((df['price']*10000000)/df['price_per_sqft']))
```

```python
df.head()
```

```python
# 1 more col added - 'property type' to distinguish between flat and plot
df.insert(loc=1, column='property_type', value='flat')
df.head()
```
